```python
# Get NFL predictions
A = pd.read_csv("~/Downloads/CHAOS #4 - Sheet1.csv")
A = A.drop(['Unnamed: 0', 'Date', 'Away \nTeam ', 'Home \nTeam', 'Enter\nYours', 'Lock In\nTime', 'Locked\nIn', 'Result', 'right side of maybe', 'diff from odds checker', 'log score', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20'], axis=1)
```

```python
A.columns = ['QBA', 'OddsChecker', 'Frelund', 'ESPN', 'Elo', 'QED']
print(A.columns)
```

```python
mycrowds = [{'oversure':1, 'averse':1, 'truth':1}, {'averse':1, 'oversure':1, 'inside':1, 'side':1, 'truth':1}, {'inside':20, 'truth':1}]
```

```python
def medboy(vector):
    try:
        m = median(vector)
    except:
        m = None
    return m
```

```python
q = []
for c in mycrowds:
    for truthness in [0.55, 0.95, 0.75]:
        univ = universe(crowd=c, p_t=truthness, str1=str(truthness))
        univ.play()
        univ.plot_rank_freq()
        univ.plot_rank_freq(score='brier')
        q.append(('ign', truthness, c, medboy(univ.time_to_99['ign'])))
        q.append(('brier', truthness, c, medboy(univ.time_to_99['brier'])))
```

```python
for e in q:
    print(e)
```

```python
r = []
for c in mycrowds:
    univ = universe(crowd=c, d_t=random.uniform, str1="Uniform(0.2,0.8)", low=0.2, high=0.8)
    univ.play()
    univ.plot_rank_freq()
    univ.plot_rank_freq(score='brier')
    r.append(('ign', "Uniform(0.2,0.8)", c, medboy(univ.time_to_99['ign'])))
    r.append(('brier', "Uniform(0.2,0.8)", c, medboy(univ.time_to_99['brier'])))
```

```python
for e in r:
    print(e)
```
